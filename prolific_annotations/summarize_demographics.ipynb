# Summarizing demographics from Prolific participants

## Load data

```python
import pandas as pd

# Load the dataset
# Adjust the path if your file is located elsewhere
file_path = 'prolific_demographic_export.csv'
df = pd.read_csv(file_path)
```

## Filter and prepare data

```python
# 1. Filter for 'AWAITING REVIEW' status
pending_df = df[df['Status'] == 'AWAITING REVIEW'].copy()

# Ensure numeric types for calculations
pending_df['Time taken'] = pd.to_numeric(pending_df['Time taken'], errors='coerce')
pending_df['Age'] = pd.to_numeric(pending_df['Age'], errors='coerce')
```

## Summarize demographics

```python
# 2. Summary Statistics for Continuous Variables
print("--- Numerical Summaries (Awaiting Review) ---")
stats_summary = pending_df[['Time taken', 'Age']].agg(['mean', 'std', 'min', 'max', 'count'])
print(stats_summary.round(2))
```

```python
# 3. Frequency Distribution for Categorical Variables
categorical_cols = [
    'Highest education level completed', 
    'Sex', 
    'Ethnicity simplified', 
    'Country of birth', 
    'Country of residence', 
    'Nationality'
]

print("--- Categorical Summaries (Counts) ---")
for col in categorical_cols:
    if col in pending_df.columns:
        print(f"\n[ {col} ]")
        counts = pending_df[col].value_counts(dropna=False)
        percentages = pending_df[col].value_counts(normalize=True, dropna=False) * 100
        
        # Combine counts and percentages for a cleaner view
        summary_df = pd.DataFrame({'Count': counts, 'Percentage (%)': percentages.round(2)})
        print(summary_df)
    else:
        print(f"\nColumn '{col}' not found in CSV.")
```
